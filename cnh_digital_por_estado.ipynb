# Número de CNHs digitais emitidas por estado, em março de 2022:

1° Tarefa: Importar a biblioteca Pandas 

```python
import pandas as pd
```

2° Tarefa: Realizar a leitura dos dados .CSV (Extraídos: https://dados.gov.br/dataset/renach/resource/61013862-1bd2-4f43-b0f8-e39a8bbe1c77)

```python
df = pd.read_csv("./cnh_digital_marco.csv", sep = ';')
print(df)
df.shape
```

3° Tarefa: Importar a biblioteca Matplotlib

```python
import matplotlib.pyplot as plt
```

4° Tarefa: Plotar o gráfico de acordo com os dados .csv

```python
x = df['Estados']
y = df['Quantidade CNH']
plt.figure(figsize=(15,5))
plt.bar(x, y)

plt.title('Quantidade de CNH emitida por estado, em março de 2022:', fontsize=15)
plt.xlabel('Estados', fontsize=15);
plt.ylabel('Quantidade CNH (mil)', fontsize=15);

plt.show()
```
